```python
import matplotlib.pyplot as plt
```

```python
reads = []
for i in range(1,6):
    with open("dead_moroz/00{}.dat".format(str(i))) as f:
        s = f.read().split('\n')
        reads.append(sorted([(float(subs.split()[0]),float(subs.split()[1])) for subs in s[1:int(s[0])+1]]))
        
reads
```

```python
x, y = map(list,zip(*reads[1]))
print(x)
print(y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
```

```python
for i in range(5):
    
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.set_aspect(1)
    x, y = zip(*reads[i])
    ax.scatter(x, y)
    
```
