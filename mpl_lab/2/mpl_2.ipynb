```python
x = []
y = []
frames = 0;
with open('data.txt') as f:
    s = f.read().split('\n')
    frames = range(int(len(s)/2))
    for i in frames:
        x.append(list(map(float,s[::2][i].split())))
        y.append(list(map(float,s[1::2][i].split()))) 
```

```python
#limsl = [(min(x[i])- 0.1*abs(min(x[i])), max(x[i])+ 0.1*abs(max(x[i])),min(y[i])- 0.1*abs(min(y[i])), max(y[i])+ 0.1*abs(max(y[i]))) for i in range(frames)]
limx = (min([(min(x[i])- 0.1*abs(min(x[i]))) for i in frames]), max([(max(x[i])+ 0.1*abs(max(x[i]))) for i in frames]))
limy = (min([(min(y[i])- 0.1*abs(min(y[i]))) for i in frames]), max([(max(y[i])+ 0.1*abs(max(y[i]))) for i in frames]))
```

```python
for i in frames:
    plt.figure()
    plt.plot(x[i], y[i])
    plt.xlim(*limx)
    plt.ylim(*limy)
    plt.title('Frame {}'.format(i+1))
    plt.grid(True)
    
```
