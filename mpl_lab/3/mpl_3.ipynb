```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("students.csv", sep=';', names=['prep', 'group', 'mark'])
df
```

```python
df[['group', 'mark']] = df[['group', 'mark']].applymap(int)
```

```python
df['prep'].value_counts()
```

```python
df['group'].value_counts()
```

```python
pr = df.groupby(['prep', 'mark']).size().unstack().fillna(0).applymap(int)
pr
```

```python
gr = df.groupby(['group', 'mark']).size().unstack().fillna(0).applymap(int)
gr
```

```python
cmap = plt.get_cmap('YlGn')
colors = cmap(np.linspace(0,1,8))
```

```python
def mark_plot(grouped, colors, title):
    fig1, ax1 = plt.subplots()
    for i in range(3,11):
        ax1.bar(grouped.index, grouped.loc[:, i].values, bottom=(grouped.loc[:, :i-1].values.sum(axis=1) if i>3 else None), label=i, color = colors[i-3])
    handles, labels = ax1.get_legend_handles_labels()
    ax1.legend(handles[::-1], labels[::-1], title='mark', loc='upper right')
    ax1.set_title(title)
```

```python
mark_plot(pr, colors, 'Marks per prep')
```

```python
mark_plot(gr, colors, 'Marks per group')
```

```python
gr_norm = gr.apply(lambda x: {i:float(x[i])/sum(x) for i in range(3,11)}, axis =1, result_type='expand')
gr_norm
```

```python
pr_norm = pr.apply(lambda x: {i:float(x[i])/sum(x) for i in range(3,11)}, axis =1, result_type='expand')
pr_norm
```

```python
mark_plot(pr_norm, colors, 'Marks per prep normed')
```

```python
mark_plot(gr_norm, colors, 'Marks per group normed')
```
