```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.pyplot import imshow
```

```python
file_path = 'imj\lunar0{}_raw.jpg'
```

```python
def ishow(im):
    arr = np.asarray(Image.fromarray(im).convert('L'))
    plt.figure()
    plt.imshow(arr, cmap='gray', vmin=0, vmax=255)
    plt.show()
```

```python
data = []
#data = np.array(np.zeros(1,3))
for i in range(1,4):
    img = Image.open(file_path.format(i))
    data.append(np.array(img, dtype=np.float64))
    ishow(np.array(img))
```

```python
pics = [np.array(d, copy = True) for d in data]
pic = pics[2]
```

```python
pics = [np.uint8((pic-np.min(pic))*(255/(np.max(pic)-np.min(pic)))) for pic in pics]
for pic in pics:
    ishow(pic)
    print('max:{}\tmin:{}'.format(np.max(pic), np.min(pic)))
```
