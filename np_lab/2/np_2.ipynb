```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'signals\signal0{}.dat'
title = ('signal 0{} raw', 'signal 0{} refined')
```

```python
data = []
for i in range(1,4):
    with open(file_path.format(i), 'r') as f:
        data.append(np.array(f.read().split(), dtype = np.float64))
```

```python
def middle(a):
    #ind = np.array([i-9 if i-9>0 else 0 in range(a.size)])
    #b = np.array(np.zeros(1,a.size))
    b = np.cumsum(a)-np.hstack((np.zeros((1,10)).flatten(), np.cumsum(np.hstack(a[:-10]))))
    b[9:] = b[9:]/10
    for i in range(9):
        b[i] = b[i]/(i+1)
    return b
```

```python
ndata = [middle(d) for d in data]
```

```python
for i in range(3):
    fig, ax = plt.subplots(1,2)
    ind = np.arange(data[i].size)
    ax[0].plot(ind, data[i])
    ax[0].set_title(title[0].format(i+1))
    ax[0].set_ylim(np.min(data[i])*0.9, np.max(data[i])*1.1)
    ax[1].plot(ind, ndata[i])
    ax[1].set_title(title[1].format(i+1))
    ax[1].set_ylim(np.min(data[i])*0.9, np.max(data[i])*1.1)
    fig.set_size_inches(15,5)
```
