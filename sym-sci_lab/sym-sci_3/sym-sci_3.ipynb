```python
import sympy as sy
import scipy as sc
import matplotlib.pyplot as plt
import numpy as np
```

**Символьное решение**

```python
x = sy.symbols('x')
y = sy.Function('y')
```

```python
ode = sy.Eq(sy.Derivative(y(x), x), -2*y(x))
```

```python
sol = sy.dsolve(ode, y(x), ics={y(0): sy.sqrt(2)})
sol
```

```python
xs = np.linspace(0, 10, 100)
ys_sym = np.array([float(sol.rhs.evalf(subs={x:i})) for i in np.linspace(0,10,100)])
```

```python
plt.plot(xs, ys_sym)
plt.grid(True)
```

**Численное решение**

```python
from scipy.integrate import odeint
```

```python
y0 = np.sqrt(2)
ys_sci = np.array(odeint(dy_dx, y0, xs)).flatten()
```

Оба решения

```python
fig, ax = plt.subplots()
ax.plot(xs, ys_sym, '-.', color='b')
ax.plot(xs, ys_sci, '--', color='r')
ax.grid(True)
fig.set_size_inches(10,10)
```
